```python
import numpy as np 
import csv
import random

n_rows = 42  # number of rows in an experimental file
condition_types = ['Male Mutant',
                   'Female Mutant',
                   'Male Wild Type', 
                   'Female  Wild Type']

# Define the mean and standard deviations for 
# each condition

mu = np.zeros(4)
sd = np.zeros(4)

# These represent Reaction Times 
# and Mutants are faster

#'Male Mutant' :
mu[0] = 12
sd[0] = 1

#'Female Mutant' :
mu[1] = 10
sd[1] = 1

#'Male Wild Type' :
mu[2] = 20
sd[2] = 1

#'Female  Wild Type' :
mu[3] = 24
sd[3] = 1

# Open the CSV file and start writing into it.
with open('018DataFile2.csv','w',newline='') as datacsv:
    # Open and start writing into the CSV file object 
    csv_obj = csv.DictWriter(datacsv, fieldnames=condition_types)
    csv_obj.writeheader() # write the header to the CSV file 
                          # (default does not)
    
    for i in range(0, n_rows) :            
        # Generate the data;
        datapoint = mu + sd * np.random.randn()
        
        #print(j)
        #datapoint
        csv_obj.writerow({'Male Mutant': mu[0] + sd[0] * np.random.randn(),
                           'Female Mutant': mu[1] + sd[1] * np.random.randn(),
                           'Male Wild Type': mu[2] + sd[2] * np.random.randn(), 
                           'Female  Wild Type': mu[3] + sd[3] * np.random.randn()})
```
